```python
# 합칙 : merge
df2 = df2.merge(df1[['id','cast','crew']], on ='id')
df2.head(2)
```

```python
# 대중들이 좋아하는 영화 축출
# 되도록 많은 사람들이 평가 (평점수도 좋아야 하지만 평가한 사람 수도 많아야 한다)
```

```python
# 일단 우선 구해야 할 값들이 있다 
#https://www.kaggle.com/code/tanusreedas/simple-recommender-system-step-by-step-explained 참조
C = df2['vote_average'].mean()
C
#영화의 전체 평균이 6점 정도 
```

```python
# 상위 10%의 데이터 뽑기
m = df2['vote_count'].quantile(0.9)
m
# 따라서 1838명 이상이 평가한 영화를 찾는 조건을 얻는다.
```

```python
# 새로운 데이터 프레임 만들기 
# 1838 이상의 사람들이 평가한 영화를 뽑기 위해
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
# 상위 10%의 데이터가 추출 되었다 481/4810
```

```python
# 졍렬 : df2 복사본 데이터의 인덱스와 평가인원수 1838 이상의 값이 정렬되어 나온다
q_movies['vote_count'].sort_values()
```

```python
# kaggle 공식 적용
def weighted_rating(x,m = m, C = C):
      v = x['vote_count']
      R = x['vote_average']
      return (v / (v+m)*R) + (m / (m +v)* C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis =1) # 데이터를 row 단위로 주기 때문에 axis = 1 로 준다 
q_movies.head(3)
```

```python
# sscore 기준으로 내림차순 정렬
q_movies = q_movies.sort_values('score',ascending=False)
q_movies[['title','vote_count','vote_average','score']].head(10)
```

```python
#칼럼에 popularity 와 비교해 보자 
# kaggle 코드 이용하여 시각화

import matplotlib.pyplot as plt
pop = df2.sort_values('popularity', ascending=False).head(10)
plt.figure(figsize=(12, 6))
bars = plt.barh(pop['title'], pop['popularity'], color='yellowgreen')
#annotations for vote average and vote count
for bar, avg, count in zip(bars, pop['vote_average'], pop['vote_count']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, 
             f'Average Vote: {avg:.2f}\nVote Count: {count}',
             va='center', ha='left', fontsize=10, color='black')
    
plt.xlabel('Popularity')
plt.ylabel('Movie Title')
plt.title('Top 10 Most Popular Movies')
plt.gca().invert_yaxis()  #Invert y-axis to show highest popularity at the top
plt.xlim(0, max(pop['popularity']) * 1.2)
plt.tight_layout()
plt.show()
```

## Content Based Filtering( 줄거리 기반 필터링)

```python
df2['overview'].head(5)
#  간략한 내용 데이터 값이 나온다 (문자열)
```

```python
# text 분석을 위한 Bag Of words - BOW (단어를 봉투에 넣어 섞는다)
# 같은 단어들이 쓰인 횟수 값을 피쳐 백터화
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words ='english')
#TfidfVectorizer 에서는 문장의 관사나, 흔히 쓰는 일반 텍스트는 제외 시켜주는 기능을 준다 (stop_words)
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
ENGLISH_STOP_WORDS
# 즉 지금 나오는 결과 값의 단어들은 제외 한다는 뜻 
```

```python
df2['overview'].isnull().values.any()
# 즉 null 값이 있다 
```

```python
# 빈값을 빈 공간으로 채워준다
df2['overview'] = df2['overview'].fillna('')
```

```python
#Vectorizer TfidfVectorizer : 각 문장에서 관사나 
tfidf_matrix = tfidf.fit_transform(df2['overview'])
tfidf_matrix.shape
```

```python
tfidf_matrix
```

```python
# 단어 유사어 파악 - 코사인 유사도 : 벡터의 방향과 기울기가 중요하다
#함수 linear_kernel 이 cosine_similarity 보다 속도가 더 빠르다 
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix) # x, y 값에 각각 데이
cosine_sim
```

```python
cosine_sim_df = pd.DataFrame(cosine_sim)
cosine_sim_df.round(1)
# 1 과 유사한 값을 구한다
```

```python
# 영화를 필요한 함수를 위한 값처리 
# 제목을 입력했을때 인덱스 값을 얻는 
indices = pd.Series(df2.index, index = df2['title']).drop_duplicates()
indices
```

```python
indices['Avatar']
```

```python
df2.iloc[[0]]
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index 값을 얻기
    idx = indices[title]
    
    # 코사인 유사도 매트릭스 (cosine_sim) 에서 idx 에 해당하는 데이터를 (idx, 유사도) 형태로 얻기
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 코사인 유사도 기준으로 내림차순 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 자기 자신을 제외한 10개의 추천 영화를 슬라이싱
    sim_scores = sim_scores[1:11]
    
    # 추천 영화 목록 10개의 인덱스 정보 추출
    movie_indices = [i[0] for i in sim_scores]
    
    # 인덱스 정보를 통해 영화 제목 추출
    return df2['title'].iloc[movie_indices]
```

## 다양한 요소 기반 추천 (장르 감독 키워드)

```python
df2.loc[0,'genres']
# dic 이가 리스트로 감싸져있다
# 그러나 실제 type 은 스트링으로 되어 있다 
```

```python
type(df2.loc[0,'genres'])
```

```python
# 데이터 가공 ---> 리스트나 dic으로 바꿔주는 literal_eval () 사용
from ast import literal_eval
```

```python
# string type을 자료형으로 바꿔주기 
features = ['cast','crew','keywords','genres']
for feature in features:
    df2[feature]=df2[feature].apply(literal_eval)
```

```python
df2.loc[0, 'crew']
```

```python
# 이 정보에서 감독 정보만 빼는 함수 만들기 
# 리스트 안을 순회하면서 감독키워드를 찾으면 이름을 얻는 함수
def get_director(x):
    for i in x:
        if i['job'] =='Director':
            return i['name']
    return np.nan # 데이터가 없으면 nan 처리     
```

```python
# 새로운 컬럼 만들기 
df2['director'] = df2['crew'].apply(get_director)
df2['director']
```

```python
# # 너무 많아서 각 피쳐의의 처음 3개에서 이름만 축출해보자 
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title','cast','director','keywords','genres']].head(3)
```

```python
# 백터화를 잘 하기 위하ㅐ 띄어쓰기 없애고 소문자로 모두 바꿔주기
#띄어쓰기가 들어가면 한사람의 이름이 두 사람이 될 수 있다 - 
```

```python
def clean_data(x):
    if isinstance(x, list):
        return[str.lower(i.replace(' ','')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ',''))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'genres','director']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title','cast','director','keywords','genres']].head(3)
```

```python
# 이 데이터를 합산한다 
# 영어 한 문장인 것 처럼 만들기 
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```
