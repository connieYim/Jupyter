* 다변수 (범주/연속)

```python
'''주로 마켓팅에 쓰는 방법'''
sns.barplot(data=df1, x='channel',y='amount')
''' 이 경우 단순히 channel.value_counts() 를 하게 되면 채널에 유입된 
사람들은 많지만 반드시 수입과 연결되지는 않기 때문에 pivot table 이용'''
```

```python
p1=pd.pivot_table(data=df1, index='channel', values= 'amount',aggfunc='sum')
p1.sort_values(by='amount', ascending=False).head(10)
```

```python
df1['channel'].value_counts() # 이렇게 리스트 순서가 다를 수 있다
```

```python
top_list=p1.sort_values(by='amount', ascending=False).head(7).index.tolist()
```

```python
sns.barplot(data=df1, x='channel',y='amount', order =top_list)
```

```python
''' 어떤 채널이 매출이 좋냐?'''
sns.barplot(data=df1, x='channel',y='amount', order =top_list,
            estimator=sum,ci=None, hue='type_of_contract')
'''ci=None # confoidence interval : 연필심 '''
```

```python
sns.boxplot(data=df1, x='channel', y='amount', order=top_list)
```

* 매직 명령어 사용 : Jupyter Notebook에서 여러 옵션을 바꿔줄때 사용

```python
df1 = pd.read_csv('Data04.csv')
df1.head()
```

```python
df1.info()
```

```python
df1['주문일(datetime)'] = pd.to_datetime(df1['주문일'],format='%Y%m%d')
```

```python
df2 = df1.set_index('주문일(datetime)')
df2.head()
```

```python
# 그래프 출력 
plt.plot(df2['상품구매금액'])

# 그래프를 출력 
plt.show()
```

* 층별화를 통한 시각화 

```python
%matplotlib  inline
```

```python
# 그래프 출력 
plt.plot(df2['상품구매금액'],label='상품구매금액')

# 타이틀과 레이블 설정 
plt.xlabel('일자')
plt.ylabel('판매 금액')
plt.title('기간 별 상품 구매금액 추이 ')

# 그래프를 출력 
plt.show()
```

* 축 설정 

```python
# 그래프 사이즈 조절
plt.figure(figsize = [10,5])

# 그래프 출력 
plt.plot(df2['상품구매금액'],label='상품구매금액')

# 축 설정
plt.xlim(['2017-01-01','2018-12-31'])
plt.ylim([0,2000000])

# 타이틀과 레이블 설정 
plt.xlabel('일자')
plt.ylabel('판매 금액')
plt.title('기간 별 상품 구매금액 추이 ')

# 그래프를 출력 
plt.show()
```
