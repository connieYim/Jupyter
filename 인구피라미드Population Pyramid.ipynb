```python
df2 = pd.read_excel('합계출산율.xlsx', skiprows=2, nrows=2, index_col=0)
df2
```

```python
df2.index
```

```python
df2.loc['출생아 수']
```

```python
df2.index.values
#알수없는 유니코드가 포함되어 있다
```

```python
df2.rename(index={'출생아\xa0수':'출생아 수', '합계\xa0출산율':'합계 출산율'},
          inplace= True)#inplace 를 해야 실제 데이터에 적용된다
```

```python
df2.loc['출생아 수']
```

```python
df2.iloc[0]
```

```python
#칼럼 위치 바꾸기
```

```python
df2 = df2.T
df2.head()
```

```python
plt.plot(df2.index, df2['출생아 수'])
plt.plot(df2.index, df2['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize =(10, 7))
ax1.set_ylabel,('출생아 수 (천 명)')
ax1.bar(df2.index, df2['출생아 수'], color ='orange')

ax2 = ax1.twinx()
ax2.set_ylabel,('합계 출산율(가임여성 1명당 명)')
ax2.plot(df2.index, df2['합계 출산율'], color ='yellow')
```

```python
fig, ax1 = plt.subplots(figsize =(13, 5))
ax1.set_ylabel,('출생아 수 (천 명)')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df2.index, df2['출생아 수'], color ='orange')

ax2 = ax1.twinx()
ax2.set_ylabel,('합계 출산율(가임여성 1명당 명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0,1])

ax2.plot(df2.index, df2['합계 출산율'], color ='yellow')
```

```python
'''그래프 위에 수치 표시 '''
```

```python
fig, ax1 = plt.subplots(figsize =(13, 5))
fig.suptitle('Number of Births & Total Fertility Rate')
ax1.set_ylabel('Number of Births ')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df2.index, df2['출생아 수'], color ='orange')

for idx, val in enumerate(df2['출생아 수']):
     ax1.text(idx, val+12, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Fertility Rate')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0,1])

ax2.plot(df2.index, df2['합계 출산율'], color ='yellow', marker='o',
         ms=15, lw =5, mec ='w', mew = 3)
for idx, val in enumerate(df2['합계 출산율']):
     ax2.text(idx, val +0.05, val, ha='center')
```
