```python
import os
import pandas as pd
from docx import Document

# 사용자 이름을 사용하여 바탕화면 경로 생성
user_name = os.getlogin()
desktop_path = f'C:/Users/Connie/Desktop'

# 엑셀 파일 경로 (실제 파일 이름으로 변경)
excel_file_path = f'{desktop_path}/emails.xlsx'  # 실제 파일 경로로 변경

# 파일 존재 여부 확인
if not os.path.exists(excel_file_path):
    print(f"파일을 찾을 수 없습니다: {excel_file_path}")
else:
    # 엑셀 파일 읽기
    df = pd.read_excel(excel_file_path)

    # 이메일 컬럼 이름을 'Email'로 가정 (실제 컬럼 이름으로 변경)
    email_column_name = 'Email'  # 실제 이메일 컬럼 이름으로 변경
    if email_column_name not in df.columns:
        print(f"컬럼 '{email_column_name}'을 찾을 수 없습니다.")
    else:
        # 이메일 주소 리스트로 추출
        emails = df[email_column_name].dropna().tolist()

        # 이메일 뒤에 콤마 추가
        emails = [email + ',' for email in emails]

        # 100개 단위로 나누기
        chunks = [emails[i:i + 100] for i in range(0, len(emails), 100)]

        # 각 100개 단위로 워드 파일 저장
        for idx, chunk in enumerate(chunks):
            doc = Document()
            doc.add_paragraph(' '.join(chunk))
            output_path = os.path.join(desktop_path, f'output_chunk_{idx+1}.docx')
            doc.save(output_path)
            print(f"워드 파일 저장 경로: {output_path}")

        print(f"Emails have been successfully processed and saved on your Desktop.")
```
