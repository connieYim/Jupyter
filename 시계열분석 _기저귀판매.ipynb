```python
import statsmodels.tsa.api as tsa
```

```python
df_time = pd.pivot_table(data=df2, index='Datetime', values='구매금액', 
                                   aggfunc='sum')
df_time.head()
```

```python
'''시계열분석은 index 가 X, values가 Y'''
```

```python
 y= df_time['구매금액'].resample('1D').mean()
#일주일 단위 판매평균금액 --> 데이터 레코드 수가 50개 전도여서 모델 조건에 맞지 않음
# 하루 단위로 바꿈
```

```python
'''결측치 확인'''
y.isnull().sum()
```

```python
''' 있는 경우는 
y.fillna( method ='bfill') : 전날 데이터로 채워주기
y.fillna( method ='ffill') :forward  데이터로 채우기'''
y1 = y.fillna(method ='ffill') 
```

```python
'''시계열 분해
Trend/ Seasonality / Resdiual'''

model_series =tsa.seasonal_decompose(y1, model = 'additive')
fig = model_series.plot()
```

```python
'''시계열 분해를 통해서 데이터 트랜드를 알아본 후에 ARIMA 적용'''
```

```python
import itertools # 반복수를 만드는 라이브러리 
```

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq
```

```python
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
param_list = []
param_seasonal_list = []
results_AIC_list = []
```

```python
for param in pdq:
     for param_seasonal in seasonal_pdq:
        try:
            mod = tsa.statespace.SARIMAX(y1,order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            param_list.append(param)
            param_seasonal_list.append(param_seasonal)
            results_AIC_list.append(results.aic)
        except:
             continue
```

```python
ARIMA_list = pd.DataFrame({'Parameter':param_list,'Seasonal':param_seasonal_list,'AIC':results_AIC_list})
ARIMA_list.to_excel('arima_model_list.xlsx')
```

```python
ARIMA_list.sort_values(by='AIC')
```

```python
mod = tsa.statespace.SARIMAX(y1,order=(1,0,1),seasonal_order=(0, 1, 1, 12),
                              enforce_stationarity=False, enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
mod = tsa.statespace.SARIMAX(y1,order=(0, 1, 1),seasonal_order=(0, 1, 1, 12),
                              enforce_stationarity=False, enforce_invertibility=False)
results = mod.fit()
print(results.summary())
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
results.get_prediction()
```

```python
y1.head()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2019-04-22'), dynamic=False)
pred_ci = pred.conf_int()

ax = y1.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1], color='k', alpha=.2)

plt.legend()
plt.show()
```
